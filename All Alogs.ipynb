```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
```

```python
d = fetch_california_housing()
d
```

```python
df = pd.DataFrame(d.data,columns=d.feature_names)
df['target']=d.target
```

```python
df
```

```python
df.isnull().sum().sum()
```

```python
X = df.drop(columns=['target'])
y = df['target']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st = StandardScaler()
x_scaled = st.fit_transform(X)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

```python
x_trains,x_tests,y_trains,y_tests = train_test_split(x_scaled,y,test_size=0.2,random_state=2)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.svm import SVR
```

```python
def model_eval(model):
    model.fit(x_trains,y_trains)
    y_pred_train = model.predict(x_trains)
    y_pred_test = model.predict(x_tests)
    print('Performace on Traning Data ',r2_score(y_trains,y_pred_train))
    print('Performace on Testing Data ',r2_score(y_tests,y_pred_test))
```

```python
model_eval(LinearRegression())
```

```python
model_eval(KNeighborsRegressor())
```

```python
model_eval(SVR())
```

```python
def model_eval2(model):
    model.fit(x_train,y_train)
    y_pred_train = model.predict(x_train)
    y_pred_test = model.predict(x_test)
    print('Performace on Traning Data ',r2_score(y_train,y_pred_train))
    print('Performace on Testing Data ',r2_score(y_test,y_pred_test))
    return model.feature_importances_
```

```python
model_eval2(DecisionTreeRegressor())
```

```python
model_eval2(RandomForestRegressor())
```

```python
model_eval2(GradientBoostingRegressor())
```

```python
model_eval2(XGBRegressor())
```

```python
model_eval2(LGBMRegressor())
```
